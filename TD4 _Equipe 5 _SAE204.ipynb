```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. On charge le fichier
df = pd.read_csv('faker_output_v2/Membre.csv')

# 2. Extraction simplifiée de la ville (le texte à la fin de l'adresse)
# On récupère ce qui n'est pas un chiffre à la fin de la ligne
df['ville'] = df['adresse'].str.extract(r'(\D+)$')[0].str.strip()

# 3. On récupère les 10 villes les plus fréquentes (le Top 10)
top_10 = df['ville'].value_counts().head(10)

# 4. ON TRIE PAR ORDRE CROISSANT
# C'est cette ligne qui fait la magie : du plus petit au plus grand
top_10_croissant = top_10.sort_values(ascending=True)

# 5. Création du graphique
plt.figure(figsize=(12, 6))
top_10_croissant.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Top 10 des territoires avec le plus de membres (Ordre croissant)')
plt.xlabel('Villes / Territoires')
plt.ylabel('Nombre de membres')
plt.xticks(rotation=45, ha='right') # Inclinaison pour que les noms ne se chevauchent pas
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. On charge le fichier 'contient.csv'
contient = pd.read_csv('contient.csv')

# 2. On prend un échantillon de 100 repas uniques
repas_ids = contient['IdR'].unique()[:100]
df_100 = contient[contient['IdR'].isin(repas_ids)]

# 3. On compte le nombre de plats (IdP) par repas (IdR)
plats_par_repas = df_100.groupby('IdR')['IdP'].count()

# 4. On calcule la DISTRIBUTION
# On regarde combien de repas ont 10 plats, 15 plats, etc.
distribution = plats_par_repas.value_counts().sort_index()

# 5. Création du graphique en barres
plt.figure(figsize=(12, 6))
distribution.plot(kind='bar', color='orange', edgecolor='black')

# On ajoute les valeurs au-dessus des barres pour la précision
for i, v in enumerate(distribution):
    plt.text(i, v + 0.2, str(int(v)), ha='center', fontweight='bold')

plt.title('Distribution du nombre de plats par repas (Échantillon de 100 repas)', fontsize=14)
plt.xlabel('Nombre de plats dans le repas')
plt.ylabel('Nombre de repas')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
